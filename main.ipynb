# Data Cleaining

```python
clean_df= train_data.dropna()
```

```python
numerical_clean_df = clean_df.drop(['Destination','PassengerId','Cabin','Name','HomePlanet'],axis=1)
```

```python
##convert bool to int
numerical_clean_df[['CryoSleep','VIP','Transported']] = numerical_clean_df[['CryoSleep','VIP','Transported']].astype(int)
```

```python
numerical_clean_df
```

```python
numerical_clean_df.dtypes
```

```python
corr =  numerical_clean_df.corr()
```

```python
corr['Transported'].sort_values(ascending=False)
```

```python
k=9
cols = corr.nlargest(k,'Transported')['Transported'].index
print(cols)
cm = np.corrcoef(numerical_clean_df[cols].values.T)
f,ax = plt.subplots(figsize=(14,12))
sns.heatmap(cm,vmax=0.8,linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels=cols.values,annot_kws={'size':12},
            yticklabels=cols.values)
```
